```r
#Charger les librairies
library(Seurat)
```

```r
#Charger les objets Seurat

seurat_M15_mRNA_nofilter <- readRDS("7_Article/results/Seurat/M15_mRNA_nofilter_DO.rds")
seurat_M14_mRNA_nofilter <- readRDS("7_Article/results/Seurat/M14_mRNA_nofilter_DO.rds")
```

```r
#Test d'affichage des données
seurat_M14_mRNA_nofilter 
head(seurat_M14_mRNA_nofilter@meta.data)
dim(seurat_M14_mRNA_nofilter@meta.data)  # Vérifie le nombre de cellules et de colonnes
colnames(seurat_M14_mRNA_nofilter@meta.data)  # Liste les colonnes disponibles

```

```r

#Test d'affichage des données
seurat_M15_mRNA_nofilter 
head(seurat_M15_mRNA_nofilter@meta.data)
dim(seurat_M15_mRNA_nofilter@meta.data)  # Vérifie le nombre de cellules et de colonnes
colnames(seurat_M15_mRNA_nofilter@meta.data)  # Liste les colonnes disponibles
```

```r

# Fusionner les objets
seurat_combined <- merge(
  x = seurat_M15_mRNA_nofilter, 
  y = seurat_M14_mRNA_nofilter, 
  add.cell.ids = c("Rep1", "Rep2"), 
  project = "Combined_M14_M15_mRNA_nofilter_DO"
)

# Vérifier la structure de l'objet fusionné
print(seurat_combined)
```

```r
seurat_combined$orig.ident <- sapply(strsplit(rownames(seurat_combined@meta.data), "_"), `[`, 1)
```

```r
head(seurat_combined$orig.ident)
table(seurat_combined$orig.ident)
```

```r
#Test d'affichage des données
seurat_combined 
head(seurat_combined@meta.data, 100)
tail(seurat_combined@meta.data, 100)
dim(seurat_combined@meta.data)  # Vérifie le nombre de cellules et de colonnes
colnames(seurat_combined@meta.data)  # Liste les colonnes disponibles
```

```r
# Afficher la structure de l'objet Seurat
str(seurat_combined)
```

```r
# Accéder à la couche counts.1
counts_1 <- seurat_combined@assays$RNA@layers$counts.1

# Accéder à la couche counts.2
counts_2 <- seurat_combined@assays$RNA@layers$counts.2

# Afficher un aperçu de ces matrices de comptage
head(counts_1)
head(counts_2)
```

```r
# Fusionner les deux couches de comptage
combined_counts <- cbind(counts_1, counts_2)

# Vérifier la dimension du résultat
dim(combined_counts)


# Ajouter les données combinées sous une nouvelle couche
seurat_combined@assays$RNA@layers$combined_counts <- combined_counts

# Vérifier les couches disponibles
names(seurat_combined@assays$RNA@layers)
```

```r
# Filtrer les données pour Rep1 et trier par nCount_RNA
rep1_data <- seurat_combined@meta.data[seurat_combined$orig.ident == "Rep1", ]
top_100_rep1 <- rep1_data[order(-rep1_data$nCount_RNA), ][1:100, ]

# Afficher les 100 premiers résultats
head(top_100_rep1, 100)
```

```r
# Filtrer les données pour Rep2 et trier par nCount_RNA
rep2_data <- seurat_combined@meta.data[seurat_combined$orig.ident == "Rep2", ]
top_100_rep2 <- rep2_data[order(-rep2_data$nCount_RNA), ][1:100, ]

# Afficher les 100 premiers résultats
head(top_100_rep2, 100)
```

```r
# Trier les cellules en fonction de nCount_RNA de manière décroissante
top_100_cells <- head(seurat_combined@meta.data[order(-seurat_combined@meta.data$nCount_RNA), ], 1000)

# Afficher les 100 meilleures cellules
top_100_cells
```
