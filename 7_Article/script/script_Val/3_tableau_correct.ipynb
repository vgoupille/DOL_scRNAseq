```r
video youtube qui explique bien les étapes pour le split et merge : https://youtu.be/nSNMWdgv31Y?si=QsZu1J4TV-UqaV2Z 



library(Seurat)
library(tidyverse)
library(patchwork)

options(Seurat.object.assay.version = "v3")
options(Seurat.object.assay.version = "v5")

# Load in three GSE healthy donor sequencing data
GSE128033<- Read10X(data.dir = "../Desktop/Video_Tutorials/Data/GSE128033/")
GSE128033 <- CreateSeuratObject(counts = GSE128033, min.cells = 3, min.features = 200)
GSE128033 <- PercentageFeatureSet(GSE128033, pattern = "^MT-", col.name = "percent.mt")
VlnPlot(GSE128033, features = c("nFeature_RNA", "nCount_RNA", "percent.mt"), ncol = 3)
GSE128033 <- subset(GSE128033, subset = nFeature_RNA < 5000 & nCount_RNA < 25000 & percent.mt < 20)
GSE128033$GSE <- "GSE128033"
view(GSE128033@meta.data)

class(GSE128033[["RNA"]])
GSE128033

GSE128169<- Read10X(data.dir = "../Desktop/Video_Tutorials/Data/GSE128169/")
GSE128169 <- CreateSeuratObject(counts = GSE128169, min.cells = 3, min.features = 200)
GSE128169 <- PercentageFeatureSet(GSE128169, pattern = "^MT-", col.name = "percent.mt")
VlnPlot(GSE128169, features = c("nFeature_RNA", "nCount_RNA", "percent.mt"), ncol = 3)
GSE128169 <- subset(GSE128169, subset = nFeature_RNA < 5000 & nCount_RNA < 25000 & percent.mt < 20)
GSE128169$GSE <-"GSE128169"
view(GSE128169@meta.data)
class(GSE128169[["RNA"]])

GSE132771<- Read10X(data.dir = "../Desktop/Video_Tutorials/Data/GSE132771/")
GSE132771 <- CreateSeuratObject(counts = GSE132771, min.cells = 3, min.features = 200)
GSE132771 <- PercentageFeatureSet(GSE132771, pattern = "^MT-", col.name = "percent.mt")
VlnPlot(GSE132771, features = c("nFeature_RNA", "nCount_RNA", "percent.mt"), ncol = 3)
GSE132771 <- subset(GSE132771, subset = nFeature_RNA < 4000 & nCount_RNA < 20000 & percent.mt < 10)
GSE132771$GSE <-"GSE132771"
view(GSE132771@meta.data)

# merge into one obj
obj <- merge(x= GSE128033, y=list(GSE128169, GSE132771))
view(obj@meta.data)
class(obj[["RNA"]])

# obj.list <- SplitObject(obj, split.by = "GSE")
obj[["RNA"]] <- split(obj[["RNA"]], f = obj$GSE)
obj
class(obj[["RNA"]])
Layers(obj[["RNA"]])

obj <- NormalizeData(obj)
obj <- FindVariableFeatures(obj)
obj <- ScaleData(obj)
obj <- RunPCA(obj)
obj <- FindNeighbors(obj)
obj <- FindClusters(obj, resolution = 0.8, cluster.name = "unintegrated_clusters")
obj <- RunUMAP(obj, dims = 1:10)
p1 <- DimPlot(obj, group.by = "GSE") + ggtitle("unintegrated")
p1

# obj <- NormalizeData(obj)
# obj <- FindVariableFeatures(obj)
# obj <- ScaleData(obj)
# obj <- RunPCA(obj)
# Perform streamlined (one-line) integrative analysis
# Anchor-based CCA integration (method=CCAIntegration)
obj <- IntegrateLayers(object = obj, method = CCAIntegration, group.by = "GSE",
                       orig.reduction = "pca", new.reduction = "integrated.cca",
                       verbose = FALSE)
obj <- FindNeighbors(obj, reduction = "integrated.cca", dims = 1:30)
obj <- FindClusters(obj, resolution = 0.8, cluster.name = "cca_clusters")
obj <- RunUMAP(obj, reduction = "integrated.cca", dims = 1:30, reduction.name = "umap.cca")
p2 <- DimPlot(obj, reduction = "umap.cca", group.by = "GSE") + ggtitle("cca_integrated")
p1 + p2

# Anchor-based RPCA integration (method=RPCAIntegration)
obj <- IntegrateLayers(object = obj, method = RPCAIntegration,
                       orig.reduction = "pca", new.reduction = "integrated.rpca",
                         verbose = FALSE)
obj <- FindNeighbors(obj, reduction = "integrated.rpca", dims = 1:30)
obj <- FindClusters(obj, resolution = 0.8, cluster.name = "rpca_clusters")
obj <- RunUMAP(obj, reduction = "integrated.rpca", dims = 1:30, reduction.name = "umap.rpca")
p3 <- DimPlot(obj, reduction = "umap.rpca", group.by = "GSE") + ggtitle("rpca_integrated")
p1 + p3

obj <- JoinLayers(obj)
```
