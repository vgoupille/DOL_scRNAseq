```r
library (Seurat)
```

```r
#Charger les objets Seurat

seurat_M15_mRNA_nofilter <- readRDS("7_Article/results/Seurat/M15_mRNA_nofilter_DO.rds")
seurat_M14_mRNA_nofilter <- readRDS("7_Article/results/Seurat/M14_mRNA_nofilter_DO.rds")
```

```r

#Test d'affichage des données
seurat_M15_mRNA_nofilter 
head(seurat_M15_mRNA_nofilter@meta.data)
dim(seurat_M15_mRNA_nofilter@meta.data)  # Vérifie le nombre de cellules et de colonnes
colnames(seurat_M15_mRNA_nofilter@meta.data)  # Liste les colonnes disponibles
```

```r
head (seurat_M15_mRNA_nofilter, 100)
```

```r
# Créer un vecteur contenant la nouvelle colonne basée sur 'sublibrary'
rep_column <- ifelse(seurat_M15_mRNA_nofilter$sublibrary == "M15", "rep1",
                     ifelse(seurat_M15_mRNA_nofilter$sublibrary == "M14", "rep2"))

# Créer un data.frame avec les valeurs à ajouter
metadata_to_add <- data.frame(rep = rep_column)
rownames(metadata_to_add) <- colnames(seurat_M15_mRNA_nofilter)

# Ajouter la nouvelle métadonnée à l'objet Seurat
seurat_M15_mRNA_nofilter <- AddMetaData(seurat_M15_mRNA_nofilter, metadata = metadata_to_add)

# Vérifier si la colonne a été ajoutée
head(seurat_M15_mRNA_nofilter@meta.data)
```

```r
#supprimer la colonne orig.ident

seurat_M15_mRNA_nofilter@meta.data$orig.ident <- NULL

# Vérifier que la colonne a été supprimée
head(seurat_M15_mRNA_nofilter@meta.data)
```

```r
#modifier l'ordre des colonnes
seurat_M15_mRNA_nofilter@meta.data <- seurat_M15_mRNA_nofilter@meta.data[, c("nCount_RNA", "nFeature_RNA", "rep", "sublibrary", "well", "cond")]

# Vérifier le nouvel ordre des colonnes
head(seurat_M15_mRNA_nofilter@meta.data)
```

```r
# Créer un vecteur contenant la nouvelle colonne basée sur 'sublibrary'
rep_column <- ifelse(seurat_M14_mRNA_nofilter$sublibrary == "M15", "rep1",
                     ifelse(seurat_M14_mRNA_nofilter$sublibrary == "M14", "rep2"))

# Créer un data.frame avec les valeurs à ajouter
metadata_to_add <- data.frame(rep = rep_column)
rownames(metadata_to_add) <- colnames(seurat_M14_mRNA_nofilter)

# Ajouter la nouvelle métadonnée à l'objet Seurat
seurat_M14_mRNA_nofilter <- AddMetaData(seurat_M14_mRNA_nofilter, metadata = metadata_to_add)

# Vérifier si la colonne a été ajoutée
head(seurat_M14_mRNA_nofilter@meta.data)




#supprimer la colonne orig.ident

seurat_M14_mRNA_nofilter@meta.data$orig.ident <- NULL

# Vérifier que la colonne a été supprimée
head(seurat_M14_mRNA_nofilter@meta.data)



#modifier l'ordre des colonnes
seurat_M14_mRNA_nofilter@meta.data <- seurat_M14_mRNA_nofilter@meta.data[, c("nCount_RNA", "nFeature_RNA", "rep", "sublibrary", "well", "cond")]

# Vérifier le nouvel ordre des colonnes
head(seurat_M14_mRNA_nofilter@meta.data)
```

```r
rep1 <- seurat_M15_mRNA_nofilter
rep2 <- seurat_M14_mRNA_nofilter
```

```r

# Fusionner les objets Seurat rep1 et rep2
combined_seurat <- merge(rep1, y = rep2, add.cell.ids = c("rep1", "rep2"))

# Vérifier les couches présentes
names(combined_seurat[["RNA"]])
```
